```python
import pandas as pd
import numpy as np
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Singapore','Houston','Japan','China','Singapore', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
#check whether 'Bob', 'India' and 'M-90' is in data or not
df.isin(['Bob', 'India','M-90'])
```

```python
# Filter data according to 'Bob', 'India' and 'M-90' is in data or not
df[df.isin(['Bob', 'India','M-90']).any(axis=1)]
```

```python
#convert all string values to lowercase
df[df['Name'].str.lower().isin(['bob'])]
```
