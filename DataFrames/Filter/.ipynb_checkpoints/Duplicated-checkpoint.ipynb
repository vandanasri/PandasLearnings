```python
import pandas as pd
import numpy as np
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Los Angeles','Houston','Los Angeles','Houston','Chicago', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
#check duplicate in name
df['Name'].duplicated()
```

```python
#Duplicate Value in name
df[df['Name'].duplicated()]
```

```python
# last one is not duplicate but reset are
df['City'].duplicated(keep='last')
```

```python
df[df['City'].duplicated(keep='first')]
```

```python
df[df['City'].duplicated(keep='last')]
```

```python
df[df['City'].duplicated(keep=False)]
```
