```python
import pandas as pd
import numpy as np
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Singapore','Houston','Japan','China','Singapore', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
#check isnull
df['Name'].isnull()
```

```python
# null values
df[df['Name'].isnull()]
```

```python
#check value notnull
df['Name'].notnull()
```

```python
#show notnull values
df[df['Name'].notnull()]
```

```python
#To select all Id where 'Name' is available but 'Age' is missing
#and belongs to 'Chicago' and 'China'.
a = df['City'].isin(['Chicago','China'])
b = df['Name'].notnull()
c = df['Age'].isnull()
df[a & b & c]
```

```python
#place true where value is mising
df.isnull()
```

```python
#place true where value is not null
df.notnull()
```
