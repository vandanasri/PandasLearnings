```python
import pandas as pd
import numpy as np
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Los Angeles','Houston','Los Angeles','Houston','Chicago', np.nan],
    'Marks': [60, 80,  np.nan,89, 32, 45, 78, 56,  90]
}
df = pd.DataFrame(data)
df.set_index('Id',inplace = True)
df
```

```python
# drop columns
df.drop(columns='Age')
```

```python
# drop column with axis
df.drop('City',axis=1)
```

```python
#drop multiple columns using index
df.drop(columns=['Age','City'])
```

```python
#drop multiple column using index
df.drop(['Age','City'], axis=1)
```

```python
# Delete 'Age' column by using pop()
df.pop('Age')
df
```

```python
#delete multiple column
del df['City'],df['Name']
df
```

```python
#delete Single column
del df['Marks']
df
```
