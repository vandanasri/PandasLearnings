```python
import pandas as pd

# DataFrame with a custom index
data = {
    'A': [1, 2, 3, 4],
    'B': ['a', 'b', 'c', 'd'],
    'C': [5, 6, 7, 8]
}
df= pd.DataFrame(data).set_index('B')
```

```python
# Resetting the index
df.reset_index()
```

```python
# Resetting the index and dropping the original index column
df.reset_index(drop=True)
```

```python
#DataFrame with a MultiIndex
arrays = [
    ['bar', 'bar', 'baz', 'baz'],
    ['one', 'two', 'one', 'two']
]
index = pd.MultiIndex.from_arrays(arrays, names=['first', 'second'])
df_multi = pd.DataFrame(data, index=index)
df_multi
```

```python
# resetting multiple index
df_multi.reset_index()
```

```python
# Resetting only the 'second' level of the MultiIndex
df_multi.reset_index(level='second')
```

```python
# dropping multiple index
df_multi.reset_index(drop=True)
```
