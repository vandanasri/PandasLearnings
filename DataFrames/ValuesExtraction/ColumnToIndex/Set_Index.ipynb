```python
# Sample DataFrame
data = {
    'A': [1, 2, 3, 4],
    'B': ['a', 'b', 'c', 'c'],
    'C': [5, 6, 7, 8]
}
df = pd.DataFrame(data)

# Setting column 'B' as the index
df_new = df.set_index('B')
df_new
```

```python
# Setting columns 'A' and 'B' as the index
df.set_index(['A', 'B'])
```

```python
# Setting column 'B' as the index but keeping the original column A
df.set_index('B', drop=False)
```

```python
# Appending column 'B' to the existing index
df.set_index('B', append=True)
```

```python
# Trying to set column 'B' as the index with verify_integrity=True
try:
    df.set_index('B', verify_integrity=True)
except ValueError as e:
    print("Value Error: " ,e)
```
