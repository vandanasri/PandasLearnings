```python
import pandas as pd
import numpy as np
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Los Angeles','Houston','Los Angeles','Houston','Chicago', np.nan]
}
df = pd.DataFrame(data)
df.set_index('Id',inplace = True)
df
```

```python
#Fetching data with row_index 'M-90'
df.loc['M-90']
```

```python
#Fetching data with row_index 'M-90 and column_index 'Name'
df.loc['M-90','Name']
```

```python
#Fetching data with multiple row_index ['M-90','L-80'] and column_index ['Name','City']
df.loc[['M-90','L-80'],['Name','City']]
```

```python
#Fetching data only column_index ['Name','City']
df.loc[:,['Name','City']]
```

```python
#Fetching data usng slice operator
df.loc['B-30':'A-60']
```

```python
#Fetching data using slice operator
df.loc['B-30':'A-60':2]
```

```python
#Fetching data using booolean array
a = df['Age']>27
df.loc[a]
```

```python
#Fetching data using callable
b = lambda df: df['Age'] > 27
df.loc[b]
```
