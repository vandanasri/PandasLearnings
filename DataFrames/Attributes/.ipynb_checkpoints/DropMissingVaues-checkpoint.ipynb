```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Id': ['A-20','B-30','C-40','D-50','E-60','F-70','G-80','H-90'],
    'Name': ['Alice', 'Bob', 'Charlie', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Houston','Japan','China','Singapore', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
#drop all mising values row and columns
df.dropna()
```

```python
#drop row and column value with mising values
df.dropna(how = 'any')
```

```python
#If all values are mising then drop rows and columns
df.dropna(how = 'all')
```

```python
df.dropna(axis=0,how = 'all')
```

```python
#Dropping columns with all NaNs
df.dropna(axis=1,how = 'any')
```

```python
#Keep any row in the DataFrame df that has at least 1 non-null value.
df.dropna(thresh=1)
```

```python
# any row with fewer than 2 non-null values will be removed.
df.dropna(thresh=2)
```

```python
# any row with fewer than 4 non-null values will be removed.
df.dropna(thresh=4)
```

```python
dp = pd.DataFrame({'A': [1, None, 3], 'B': [None, 5, 6]})

print(dp)
df_cleaned = dp.dropna(thresh=2)

print('-----------------------------')
print(df_cleaned)
```
