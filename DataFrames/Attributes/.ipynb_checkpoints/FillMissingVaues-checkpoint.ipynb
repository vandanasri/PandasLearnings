```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Id': ['A-20','B-30','C-40','D-50','E-60','F-70','G-80','H-90'],
    'Name': ['Alice', 'Bob', 'Charlie', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Houston','Japan','China','Singapore', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
# replace NaN value with 'Sorry'
df.fillna('Sorry')
```

```python
#df['Name','Age','City'].fillna('No Name',0,'No City')
df['Name'].fillna('No Name')
```

```python
#Replace a NaN value of column 'Name' and assign to column 'Name'.
df['Name'] = df['Name'].fillna('No Name')
df
```

```python
#if I don't want to assign to df['Name'] then use inplace=True
df['Name'].fillna('No Name', inplace=True)
df
```

```python
#assigning value to age by replacing NaN and convertingit from float to int
df['Age'] = df['Age'].fillna(0).astype('int')
df
```

```python
# using category type
df['Name'] = df['Name'].astype('category')
df
```
