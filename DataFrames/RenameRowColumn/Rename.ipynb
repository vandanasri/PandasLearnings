```python
import pandas as pd
import numpy as np
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Los Angeles','Houston','Los Angeles','Houston','Chicago', np.nan]
}
df = pd.DataFrame(data)
df.set_index('Id',inplace = True)
df
```

```python
#rename index using mapper and default axis=0
df.rename(mapper={'H-20':'Id-220', 'K-90':'Id-290'})
```

```python
#rename index using index and default axis=0
df.rename(index={'M-90':'Id-190', 'A-60':'Id-250'})
```

```python
#rename column using columns
df.rename(columns={'Name':'Candidate Name', 'Age':'Candidate Age'})
```

```python
#rename columns and default axis=1
df.rename({'Name':'candidate_name', 'Age':'candidate_age'},axis=1)
```

```python
# Rename column by using columns attribute
df.columns =['cad_name','cad_age','cad_city']
df
```
