```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Singapore','Houston','Japan','China','Singapore', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
# Sorting default index
df.sort_index()
```

```python
# Sorting default index in reverse order
df.sort_index(ascending=False)
```

```python
#Adding index and name and sorting.
dp = pd.DataFrame(data, index=['p','a','z','y','t','b','w','s','e'])
dp.index.name = 'Index_Name'
dp
```

```python
#Sort column index axis =1
dp.sort_index(axis=1)
```

```python
#Sort row index axis =0
dp.sort_index(axis=0)
```

```python
#creating multiple index
array = [['p','a','z','y','t','b','w','s','e'],[2,6,4,8,7,1,5,9,3]]
multi_index = pd.MultiIndex.from_arrays(array, names=('first_index', 'second_index'))
multi_index
```

```python
#creating DataFrame with multiple index
dpf = pd.DataFrame(data, multi_index)
dpf
```

```python
# Sorting by the 'second' level
dpf.sort_index(level='second_index')
```

```python
# Sorting by the 'first' level
dpf.sort_index(level='first_index')
```
