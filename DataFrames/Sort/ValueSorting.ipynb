```python
import pandas as pd
import numpy as np
```

```python
data = {
    'Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Singapore','Houston','Japan','China','Singapore', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
# sort values of Id 
df.sort_values(by = 'Id')
```

```python
# sort values of Id in reverse order
df.sort_values('Id', ascending = False)
```

```python
# sort values of Age, always NaN placed at last
df.sort_values('Age')
```

```python
# to bring NaN position to first
df.sort_values('Age', na_position='first')
```

```python
# sort based on multiple columns
df.sort_values(by = ['Name','City'], na_position='first')
```

```python
df.sort_values(by = ['Name','City'], ascending = [True,False])
```
