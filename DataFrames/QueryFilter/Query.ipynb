```python
import pandas as pd
import numpy as np
data = {
    'Candidate Id': ['H-20','B-30','E-40','M-90','D-50','A-60','F-70','L-80','K-90'],
    'Name': ['Alice', 'Bob', 'Charlie','Bob', np.nan,'Anny', 'Bobby', 'Charter',  np.nan],
    'Age': [24, 27,  np.nan,89, 32, 45, 33, 44,  np.nan],
    'City': [ np.nan, 'Los Angeles', 'Chicago', 'Singapore','Houston','Japan','China','Singapore', np.nan]
}
df = pd.DataFrame(data)
df
```

```python
# using query method fetch only 'BoB'data
df.query('Name == "Bob"')
```

```python
# get data where age is greater than 33
df.query('Age >33')
```

```python
# get multiple data
df.query('Age >33 | Name == "Bob" & City == "Singapore"')
```

```python
#@ symbol is used to refer to Python variables within the query string
threshold = 32
df.query('Age > @threshold')
```

```python
df.query('`Candidate Id` == "E-40"')
```

```python
#Replace column name by putting '_'
df.columns = [col_name.replace(' ','_') for col_name in df.columns]
df
```
